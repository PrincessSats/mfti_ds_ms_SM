```python
cuisine_count = {}

# Заполняем словарь, проходя по списку рецептов
for recipe in recipes:
    cuisine = recipe['cuisine']
    if cuisine in cuisine_count:
        cuisine_count[cuisine] += 1
    else:
        cuisine_count[cuisine] = 1

# Определяем кухню с максимальным количеством рецептов
max_cuisine = max(cuisine_count, key=cuisine_count.get)

# Результаты
print("Количество рецептов для каждой кухни:", cuisine_count)
print("Кухня с максимальным количеством рецептов:", max_cuisine)
print("Количество рецептов:", cuisine_count[max_cuisine])
```

```python
# Импортируем модуль json
import json 
# Импортируем функцию pprint()
from pprint import pprint 
# Импортируем модуль pandas
import pandas as pd 
# Открываем файл и связываем его с объектом "f"
with open('recipes.json') as f: 
    # Загружаем содержимое открытого файла в переменную recipes
    recipes = json.load(f) 
# Создаём объект DataFrame из списка recipes
df = pd.DataFrame(recipes) 
# Выводим на экран первые строки полученного DataFrame
display(df.head())
```

```python
len(df)
```

```python
def check_ingredient(recipe_ingredients, ingredient):
    """Возвращает 1, если ingredient есть в recipe_ingredients, иначе возвращает 0."""
    return 1 if ingredient in recipe_ingredients else 0
```

```python
# Создание реестра уникальных ингредиентов
unique_ingredients = set()
for ingredients_list in df["ingredients"]:
    unique_ingredients.update(ingredients_list)

# Преобразуем уникальные ингредиенты в отсортированный список (для удобства)
unique_ingredients = sorted(unique_ingredients)

# Добавляем столбцы для каждого уникального ингредиента и заполняем их
for ingredient in unique_ingredients:
    df[ingredient] = df["ingredients"].apply(lambda x: check_ingredient(x, ingredient))

df
```

```python
# Добавляем столбец "num_ingredients" с количеством ингредиентов для каждого рецепта
df['ingredients'] = df['ingredients'].apply(len)
```

```python
# Создаём DataFrame, читаем данные из файла в переменную df
df = pd.read_csv('recipes.csv')
```

```python
ids = list(df["id"].drop_duplicates())
ingredients = list(df.columns[3:])

def make_list(row):
  ind = []
  for ingredient, value in row.items():
      if value == 1:
          ind.append(ingredient)
  return ind

# Create empty list for storing the final structure
new_recipes = [] 

# Loop through each ID
for current_id in ids: 
  # Get the cuisine for current ID
  cuisine = df[df['id'] == current_id]['cuisine'].iloc[0] 
  
  # Get ingredients for current recipe - only look at ingredient columns
  current_ingredients = make_list(df[df['id'] == current_id][ingredients].iloc[0]) 
  
  # Create current dictionary
  current_recipe = {
      'cuisine': cuisine, 
      'id': int(current_id), 
      'ingredients': current_ingredients
  } 
  
  # Add created dictionary to the list
  new_recipes.append(current_recipe) 
```

```python
# Импорт модуля json
import json 
# Функция dumps() модуля json сериализирует объект Python в строку формата JSON. 
new_recipes = json.dumps(new_recipes) 

# Откроем файл new_recipes.json для записи
with open("data/new_recipes.json", "w") as write_file: 
    # Записываем содержимое подготовленные данные в файл
    write_file.write(new_recipes)
```

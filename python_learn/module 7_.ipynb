```python
def outliers_iqr(data, feature):
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_iqr(sber_data, 'full_sq')
print(f'Число выбросов по методу Тьюки: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```

```python
def outliers_iqr_mod(data, feature, left=1.5, right=1.5):
  x = data[feature]
  quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75)
  iqr = quartile_3 - quartile_1
  lower_bound = quartile_1 - (iqr * left)
  upper_bound = quartile_3 + (iqr * right)
  outliers = data[(x < lower_bound) | (x > upper_bound)]
  cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
  return outliers, cleaned

# Load the test data
test_data = sber_data.copy()


# Test the function on a feature, e.g., 'one'
outliers, cleaned_data = outliers_iqr_mod(test_data, 'full_sq', left=1, right=6)
outliers.value_counts().sum()
```

```python
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 3 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

sber_copy = sber_data.copy()
```

```python
df = sber_data.copy()

# Step 1: Clean the data
# Remove columns with more than 50% missing values
threshold = len(df) * 0.5
df = df.dropna(axis=1, thresh=threshold)

# Remove rows with more than two missing values
df = df.dropna(thresh=len(df.columns) - 2)

# Fill missing values
for column in df.columns:
    if df[column].dtype in ['float64', 'int64']:  # Numerical columns
        df[column].fillna(df[column].mean(), inplace=True)
    else:  # Categorical columns
        df[column].fillna(df[column].mode()[0], inplace=True)

# Step 2: Plot the histogram for 'price_doc'
# Assuming 'price_doc' is a column in the DataFrame
if 'price_doc' in df.columns:
    # Calculate log of price_doc
    log_price = np.log(df['price_doc'])

    # Calculate mean and standard deviation
    mean_log_price = log_price.mean()
    std_log_price = log_price.std()

    # Plot histogram
    plt.figure(figsize=(10, 6))
    plt.hist(log_price, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

    # Add mean line
    plt.axvline(mean_log_price, color='red', linestyle='--', linewidth=2, label='Mean')

    # Add three-sigma lines
    plt.axvline(mean_log_price + 3 * std_log_price, color='green', linestyle='--', linewidth=2, label='Mean + 3σ')
    plt.axvline(mean_log_price - 3 * std_log_price, color='green', linestyle='--', linewidth=2, label='Mean - 3σ')

    # Add labels and title
    plt.title('Logarithmic Histogram of Price_doc', fontsize=16)
    plt.xlabel('Log(Price_doc)', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.legend()

    # Show plot
    plt.tight_layout()
    plt.show()
```

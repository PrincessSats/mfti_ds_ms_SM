```python
# Граф представлен в виде списка смежности (словарь)
graph = {
    1: [2, 5, 4],
    2: [1, 4, 5, 6, 3],
    3: [2, 5, 6],
    4: [1, 2, 5, 7, 8],
    5: [1, 2, 3, 4, 6, 8],
    6: [2, 3, 5, 8],
    7: [4, 8],
    8: [4, 8, 5, 6]
}

# Множество для хранения посещённых вершин
visited = set()

def dfs(vertex):
    # Помечаем текущую вершину как посещённую
    print(f"Посещаем вершину {vertex}")
    visited.add(vertex)
    
    # Проходим по всем соседям текущей вершины
    for neighbor in graph[vertex]:
        if neighbor not in visited:
            dfs(neighbor)

# Запуск DFS с вершины 1
dfs(1)
```

```python
from collections import deque

def bfs(start):
    # Очередь для хранения вершин, которые нужно посетить
    queue = deque([start])
    
    # Множество для хранения посещённых вершин
    visited = set([start])
    
    # Пока очередь не пуста
    while queue:
        # Извлекаем вершину из начала очереди
        vertex = queue.popleft()
        print(f"Посещаем вершину {vertex}")
        
        # Проходим по всем соседям текущей вершины
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

# Запускаем BFS с вершины 1
bfs(1)
```

```python
# Define the symbols
x, y = sp.symbols('x y')

# Define the function
u = x**2 + 3*x*y + 4*y**2

# Calculate the partial derivatives
u_x = sp.diff(u, x)
u_y = sp.diff(u, y)

# Evaluate the derivatives at the point (1, 1)
u_x_value = u_x.subs({x: 1, y: 1})
u_y_value = u_y.subs({x: 1, y: 1})

# Print the results
print(f"u_x(1, 1) = {u_x_value}")
print(f"u_y(1, 1) = {u_y_value}")
```

```python
# Define the symbol
x, y = sp.symbols('x y')

# Define the function
f = x**3 + y**3 - 3*x*y

# Calculate the first derivative
f_prime = sp.diff(f, x)

# Find critical points by solving f'(x) = 0
critical_points = sp.solve(f_prime, x)

# Calculate the second derivative
f_double_prime = sp.diff(f_prime, x)

# Evaluate the second derivative at each critical point to determine the nature
extrema = {}
for point in critical_points:
  second_derivative_value = f_double_prime.subs(x, point)
  if second_derivative_value > 0:
      extrema[point] = 'local minimum'
  elif second_derivative_value < 0:
      extrema[point] = 'local maximum'
  else:
      extrema[point] = 'inconclusive'

# Print the results
print("Critical points and their nature:")
for point, nature in extrema.items():
  print(f"x = {point}: {nature}")
```

```python

# Define the symbols
x, y = sp.symbols('x y')

# Define the function
z = x**3 + y**3 - 3*x*y

# Calculate the first partial derivatives
z_x = sp.diff(z, x)
z_y = sp.diff(z, y)

# Find the critical points by setting the first partial derivatives to zero
critical_points = sp.solve((z_x, z_y), (x, y), dict=True)

# Calculate the second partial derivatives
z_xx = sp.diff(z_x, x)
z_yy = sp.diff(z_y, y)
z_xy = sp.diff(z_x, y)

# Construct the Hessian matrix
H = sp.Matrix([[z_xx, z_xy], [z_xy, z_yy]])

# Evaluate the Hessian matrix at each critical point
for point in critical_points:
    x_val = point[x]
    y_val = point[y]
    H_at_point = H.subs({x: x_val, y: y_val})
    determinant = H_at_point.det()
    f_xx_at_point = z_xx.subs({x: x_val, y: y_val})
    
    if determinant > 0:
        if f_xx_at_point > 0:
            nature = "local minimum"
        else:
            nature = "local maximum"
    elif determinant < 0:
        nature = "saddle point"
    else:
        nature = "inconclusive"
    
    print(f"Critical point: ({x_val}, {y_val}), Nature: {nature}")
```

```python
import numpy as np
 
def func_rosen(x):
    r = np.sum(100*(x[1:]-x[:-1]**2)**2 + (1-x[:-1])**2, axis=0)
    return r
```

```python
import matplotlib.pyplot as plt
graph = plt.figure(figsize=[20, 10])
ax = graph.add_subplot(projection='3d')
ax.view_init(45, 30)
x1 = np.arange(-3, 3, 0.2)
x2 = np.arange(-2, 4, 0.2)
x1, x2 = np.meshgrid(x1, x2)
y = func_rosen(np.array([x1,x2]))
surf = ax.plot_surface(x1, x2, y)
plt.show()
```

```python
from scipy.optimize import minimize
x1 = np.array([2.4, 1.5, 2.1, 0.7, 1.1])
result_point = minimize(func_rosen, x1, method='powell')
print(result_point.x)
```

```python
from scipy.optimize import minimize
x1 = np.array([2.4, 1.5, 2.1, 0.7, 1.1])
result_point = minimize(func_rosen, x1, method='nelder-mead')
print(result_point.x)
```

```python
from scipy.optimize import least_squares
x1 = np.array([2.4, 1.5, 2.1, 0.7, 1.1])
result_point = least_squares(func_rosen, x1, method='trf')
print(result_point.x)
```

```python
# Импортируем все необходимые функции из библиотеки SymPy
from sympy import init_printing, solve, symbols

# Подключаем аккуратный вывод формул
init_printing()

# Создаём переменные
x1, x2, x1_tilde, lambda1, lambda2 = symbols(
    "x_1 x_2 xtilde_2 \lambda_1 \lambda_2", real=True
)

# Пишем систему уравнений для переменных
dLdx1 = -x2 + lambda1 - lambda2
dLdx2 = -x1 + lambda1
dLdx1_tilde = 2 * lambda2 * x1_tilde
dLdlambda1 = x1 + x2 - 10
dLdlambda2 = -x1 + 6 + x1_tilde**2

# Решаем систему уравнений с помощью функции solve
solution = solve(
    [dLdx1, dLdx2, dLdx1_tilde, dLdlambda1, dLdlambda2],
    x1,
    x2,
    x1_tilde,
    lambda1,
    lambda2,
)

print(solution)
```

```python
def fun(x, y, a=1, b=1):
    return a * (x ** 2) + b * (y ** 2)
```

```python
def grad(x, y, a=1, b=1):
    return np.array([2 * a * x, 2 * b * y])
```

```python
def grad_descend(grad, step_size=0.2, num_steps=30):
    lst = []
    x = np.random.uniform(0, 3, size = 2)
    lst.append(x)
    for i in range(num_steps):
        x = x - step_size * grad(lst[-1][0], lst[-1][1])
        lst.append(x)
    return np.array(lst)
```

```python
def plot_grad(fun, trace):
    fig = plt.figure(figsize=(10, 8))
    x_ = np.linspace(-1, 1, 100)
    y_ = np.linspace(-1, 1, 100)
    x, y = np.meshgrid(x_, y_)
    z = fun(x, y)
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.6)
    ax.contour(x, y, z, zdir='z', offset=z.min())
    z_trace = fun(trace[:, 0], trace[:, 1])
    ax.plot(trace[:, 0], trace[:, 1], z_trace, "o-")
    ax.set_xlim(x.min(), x.max())
    ax.set_ylim(y.min(), y.max())
    ax.set_zlim(z.min(), z.max())
```

```python
trace = grad_descend(grad, 0.1, 20)
plot_grad(fun, trace = trace)
```
